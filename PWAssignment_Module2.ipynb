Loops are used to execute a block of code repeatedly as long as a specified condition is true or for a fixed number of iteration.

1. for loop
2. while loop

The for loop is use to iterate over a sequence and execute a block of code item in the sequence


```
# for loop
number = [1,2,3,4]
for num in  number:
    print(num)

# using range funtion

for i in range(0, 10):
    print(f"iteration {i}")



#While Loop

count = 1
while count <=5:
    print(f"count is {count}")
    count +=1

# Nested loop
for i in range(1,4):
  for j in range(1,5):
    print(f" i ={i}, j={j}")
```
